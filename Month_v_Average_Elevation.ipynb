```
from google.colab import files
uploaded = files.upload()

import io
import pandas
data = pandas.read_csv(io.BytesIO(uploaded['17-Jul-2019-T_CAPmeta.csv']))
```

```
data
```

```
i = 1
list = []
while i <13:
  list.append(data.loc[data['MONTH'] == i]['ELEVATION_m_agl'].mean())
  i += 1

print(list)
```

```
import matplotlib as mpl
from matplotlib import pyplot as plt #pyplot is a subpackage of matplotlib, and provides the interface that we'll use to create plots
plt.style.use('seaborn-whitegrid') # set default visual appearance
import numpy as np
import pandas as pd
```

```
months = range(1,13)
y_pos = np.arange(len(months))

plt.bar(y_pos, list, align='center', color = 'blue')
plt.xticks(y_pos, months)
plt.xlabel('Month')
plt.ylabel('Average Elevation')
plt.title('Average Elevation every Month')

plt.show()
```
